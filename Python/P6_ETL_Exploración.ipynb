```python
# Verificar las columnas del DataFrame
print("Columnas en el DataFrame:", df.columns)
```

```python
# Total de duplicados por columna
duplicate_counts = {}

# Iterar sobre cada columna en el DataFrame
for column in df.columns:
    # Contar los valores únicos y sus frecuencias
    counts = df[column].value_counts()
    # Filtrar los valores que aparecen más de una vez y sumar
    duplicate_counts[column] = (counts - 1)[counts > 1].sum()

# Imprimir los resultados
print("Duplicados totales por columna:")
for column, count in duplicate_counts.items():
    print(f"{column}: {count}")
```

```python
import pandas as pd

# Diccionario para almacenar los resultados de los conteos de duplicados para cada columna
duplicate_counts = {}

# Lista de columnas a verificar
columns_to_check = [
    'product_id', 'order_id', 'customer_ID', 'row_id', 'product_name',
    'category', 'city', 'country', 'discount', 'market', 'unknown',
    'order_date', 'order_priority', 'profit', 'quantity', 'region',
    'sales', 'segment', 'ship_date', 'ship_mode', 'shipping_cost',
    'state', 'sub_category', 'year', 'market2'
]

# Iterar sobre cada columna para contar duplicados
for column in columns_to_check:
    # Contar las ocurrencias de cada valor
    value_counts = df[column].value_counts()
    # Filtrar los valores que aparecen más de una vez
    duplicate_values = value_counts[value_counts > 1]
    # Sumar el total de duplicados (restar la cantidad de valores para no contar la primera ocurrencia)
    duplicate_counts[column] = duplicate_values.sum() - duplicate_values.count()

# Convertir el diccionario en un DataFrame para un mejor formato de salida
duplicates_df = pd.DataFrame(list(duplicate_counts.items()), columns=['Column', 'Duplicate Count'])

# Ordenar los resultados por nombre de columna
duplicates_df.sort_values(by='Column', inplace=True)

# Mostrar el DataFrame de duplicados
print(duplicates_df)
```

3.- Valores fuera del alcance


```python
#Outliers varibles númericas
import pandas as pd

# Cargar el DataFrame
df = pd.read_csv('superstore.csv')

# Identificar outliers en una columna numérica, por ejemplo 'sales'
Q1 = df['sales'].quantile(0.25)
Q3 = df['sales'].quantile(0.75)
IQR = Q3 - Q1

# Definir límites para determinar outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filtrar los outliers en la columna 'sales'
outliers_sales = df[(df['sales'] < lower_bound) | (df['sales'] > upper_bound)]
print("Outliers en ventas:")
print(outliers_sales)

```

```python
import pandas as pd

# Variables numéricas
numerical_columns = ['sales', 'profit']  # Asumiendo que 'sales' y 'profit' son numéricas

for column in numerical_columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    outliers = df[(df[column] < lower_bound) | (df[column] > upper_bound)]
    print(f"Outliers en {column}:")
    print(outliers)

# Variables categóricas
categorical_columns = ['category', 'city']  # Asumiendo que son categóricas

for column in categorical_columns:
    category_counts = df[column].value_counts()
    rare_categories = category_counts[category_counts < 0.01 * len(df)]
    print(f"Categorías raras en {column}:")
    print(rare_categories)

```

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd

# url de la página web a «escrapear»
url = 'https://en.wikipedia.org/wiki/List_of_supermarket_chains'

# pasar "User-agent" para simular interacción con la página usando Navegador web
headers = {"User-agent": 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/79.0.3945.130 Safari/537.36'}

respuesta = requests.get(url, headers=headers)

# El código de respuesta <200> indicará que todo salió bien
print(respuesta)
```

```python
#Guardar todas las tablas encontradas
all_tables = pd.read_html(respuesta.content, encoding = 'utf8')
```

```python
#Tablas de la web
print(f'Total de tablas encontradas: {len(all_tables)}')
```

```python
#Buscar la tabla que coincide con algun elemento de la tabla
matched_table = pd.read_html(respuesta.text, match='Company')

# imprime numero de tablas que coinciden con parametro match
print(f'Total de tablas encontradas: {len(matched_table)}')
```

```python
# Guardar tabla en variable con nombre semántico
multinacional = matched_table[0]

# Verificamos si es la tabla que buscamos
multinacional.tail(5)
```

```python
# Eliminar la última columna de un DataFrame
multinacional.drop(multinacional.columns[-1], axis=1, inplace=True)

# Verificar si se eliminó la última columna
print(multinacional.head())
```

```python
multinacional.set_index('Company', inplace = True)

# Verificamos el cambio de índice
multinacional.head(2)
```

```python
#Remover espacio entre company y las demás columnas
from unicodedata import normalize
def remove_whitespace(x):
    """Funcion para normalizar datos con Unicode para luego quitar los espacios usando .replace().

    Argumentos de entrada: Nombre de columna o lista con nombres de columnas.
    Retorna: columna o columnas sin espacios en blanco
    """
    if isinstance(x, str):
        return normalize('NFKC', x).replace(' ', '')
    else:
        return x
```

```python
# Guardamos en variable nombre de columnas a quitar espacios en blanco
numeric_cols = ['Headquarters',	'Served countries (besides the headquarters)',	'Map',	'Number of locations',	'Number of employees']

# Aplicar función remove_whitespace a columnas en variable y las reemplazamos en tabla
multinacional[numeric_cols] = multinacional[numeric_cols].apply(remove_whitespace)

# Verificamos si se quitaron los espacios en blanco
multinacional.head()
```

```python
multinacional.dtypes
```

```python
# Aseguramos que las columnas 'Number of locations' y 'Number of employees' estén en formato string
multinacional['Number of locations'] = multinacional['Number of locations'].astype(str)
multinacional['Number of employees'] = multinacional['Number of employees'].astype(str)

# Usamos una expresión regular para extraer solo el primer número en la columna 'Number of locations' y 'Number of employees'
multinacional['Number of locations'] = multinacional['Number of locations'].str.extract(r'(\d[\d,.]*)')
multinacional['Number of employees'] = multinacional['Number of employees'].str.extract(r'(\d[\d,.]*)')

# Convertimos los valores a tipo float
multinacional['Number of locations'] = multinacional['Number of locations'].str.replace(',', '').astype(float)
multinacional['Number of employees'] = multinacional['Number of employees'].str.replace(',', '').astype(float)

# Verificamos los cambios
print(multinacional[['Number of locations', 'Number of employees']].head())
```

```python
# Luego convertimos los valores a tipo float
convert_dict = {
    'Headquarters': 'string',
    'Served countries (besides the headquarters)': 'string',
    'Number of locations': 'float',
    'Number of employees': 'float'
}

# Intentamos convertir los datos de nuevo
multinacional = multinacional.astype(convert_dict)

# Verificamos los tipos de datos
print(multinacional.dtypes)
```
